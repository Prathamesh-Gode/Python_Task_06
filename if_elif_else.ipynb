# if-elif-else Examples

```python
age = 20
if age < 18:
    print("Minor")
elif age < 60:
    print("Adult")
else:
    print("Senior")
```

```python
marks = 75
if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

```python
num = -5
if num > 0:
    print("Positive")
elif num == 0:
    print("Zero")
else:
    print("Negative")
```

```python
temp = 30
if temp < 15:
    print("Cold")
elif temp < 30:
    print("Warm")
else:
    print("Hot")
```

```python
n = 7
if n % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
password = "admin123"
if password == "admin123":
    print("Access granted")
else:
    print("Access denied")
```
