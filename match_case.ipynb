# match-case Examples


```python
day = 3
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Invalid")
```

```python
signal = "red"
match signal:
    case "red":
        print("Stop")
    case "yellow":
        print("Wait")
    case "green":
        print("Go")
```

```python
choice = 2
match choice:
    case 1:
        print("Pizza")
    case 2:
        print("Burger")
    case 3:
        print("Pasta")
```

```python
grade = "A"
match grade:
    case "A":
        print("Excellent")
    case "B":
        print("Good")
    case _:
        print("Average")
```

```python
browser = "chrome"
match browser:
    case "chrome":
        print("Google Chrome")
    case "firefox":
        print("Mozilla Firefox")
    case _:
        print("Other")
```

```python
num = 1
match num:
    case 1:
        print("One")
    case 2:
        print("Two")
    case _:
        print("Unknown")
```
